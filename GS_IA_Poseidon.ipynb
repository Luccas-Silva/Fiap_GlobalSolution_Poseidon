## **Modelo de Machine Learning**

```python
# Leitura de dados CSV
dados_1producao = pd.read_csv(os.path.join('1- producao-de-plastico-global.csv'))
dados_2participacao = pd.read_csv(os.path.join('2- participacao-despejo-residuo-plastico.csv'))
dados_3destino = pd.read_csv(os.path.join('3- destino-plastico.csv'))
dados_4desperdicio = pd.read_csv(os.path.join('4- desperdicio-plastico-per-capita.csv'))
dados_5poluicao = pd.read_csv(os.path.join('5- poluicao-agua-cidades.csv'))
```

```python
# Renomeando as colunas
dados_1producao.columns = ['pais', 'ano', 'producao_plastico']
dados_1producao['codigo'] = 'OWID_WRL'
dados_2participacao.columns = ['regiao', 'codigo', 'ano', 'participacao_despejo']
dados_3destino.columns = ['pais', 'codigo', 'ano', 'reciclagem', 'queima', 'descarte_incorreto', 'aterros']
dados_4desperdicio.columns = ['pais', 'codigo', 'ano', 'desperdicio_per_capita']
dados_5poluicao.columns = ['cidade', 'regiao', 'pais', 'qualidade_ar', 'poluicao_agua']
dados_5poluicao['ano'] = 2020
```

```python
df_combinado = pd.merge(dados_1producao, dados_2participacao, on=['codigo', 'ano'], how='outer')
df_combinado = pd.merge(df_combinado, dados_3destino, on=['codigo', 'ano'], how='outer')
df_combinado = pd.merge(df_combinado, dados_4desperdicio, on=['codigo', 'ano'], how='outer')
df_combinado = pd.merge(df_combinado, dados_5poluicao, on=['pais', 'ano'], how='outer')

df_combinado.fillna(0, inplace=True)

df_combinado.rename(columns={'regiao_x': 'regiao', 'regiao_y': 'regiao'}, inplace=True)
```

```python
ohe = OneHotEncoder(handle_unknown='ignore')

df_combinado['pais'] = df_combinado['pais'].astype(str)
df_combinado['regiao'] = df_combinado['regiao'].astype(str)
df_combinado['codigo'] = df_combinado['codigo'].astype(str)
df_combinado['cidade'] = df_combinado['cidade'].astype(str)
```

```python
X_encoded = ohe.fit_transform(df_combinado[['pais', 'regiao', 'codigo', 'cidade']]).toarray()
X = pd.concat([df_combinado[['producao_plastico', 'participacao_despejo', 'reciclagem', 'queima', 'descarte_incorreto', 'aterros', 'desperdicio_per_capita']], pd.DataFrame(X_encoded)], axis=1)

features_relevantes = ['producao_plastico', 'participacao_despejo', 'reciclagem', 'queima', 'descarte_incorreto', 'aterros', 'desperdicio_per_capita']
X = X[features_relevantes]

X_train, X_test, y_train, y_test = train_test_split(X, df_combinado['qualidade_ar'], test_size=0.2, random_state=42)

modelo = LinearRegression()
modelo.fit(X_train, y_train)

y_pred = modelo.predict(X_test)

rmse = mean_squared_error(y_test, y_pred, squared=False)
print(f"RMSE: {rmse}")
```

```python
X_cenario1 = X_test.copy()
X_cenario1['producao_plastico'] = X_test['producao_plastico'] * 1.5
previsoes_cenario1 = modelo.predict(X_cenario1)

resultados_cenarios = pd.DataFrame({
    'Cenário Original': y_test,
    'Cenário 1 (Aumento na Produção)': previsoes_cenario1
})

resultados_cenarios.plot(kind='line', figsize=(10, 6))
plt.title('Cenários de Poluição e Qualidade da Água')
plt.ylabel('Qualidade da Água')
plt.xlabel('Cenários')
plt.show()
```
